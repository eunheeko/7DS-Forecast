### Create mjd/fiberid tables for each plate

```python
plates = os.listdir('/data6/SDSSspec')
plates.sort()

plates = plates[:-1]
print(f'total {len(plates)} plates')
```

```python
# create supplementary directory

if not os.path.exists('/data6/SDSSspec_sup/'):
    os.makedirs('/data6/SDSSspec_sup/')
    print('supplementary directory is created')
else:
    print('already exists')
```

```python
# create individual plate directory

for pl in plates:
    
    os.makedirs('/data6/SDSSspec_sup/' + pl)
```

```python
# create info table for specPhotoAll matching

for pl in tqdm.tqdm(plates[22:]):

    files = os.listdir('/data6/SDSSspec/' + pl)
    
    mjds = []
    fiber_ids = []

    for file in files:

        f = fits.open('/data6/SDSSspec/' + pl + '/'+ file)
        hdr = f[0].header

        mjds.append(hdr['MJD'])
        fiber_ids.append(hdr['FIBERID'])

        f.close()
        
    tbl = Table([mjds, fiber_ids], names = ('mjd_my', 'fiberid_my'))
    tbl.write('/data6/SDSSspec_sup/'+ pl + '/plate_' + pl + '.csv', overwrite = True)
```
