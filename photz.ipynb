```python
import matplotlib.pyplot as plt

from astropy.io import ascii

import numpy as np
```

```python
zs = ascii.read('./output/photz.zout')
```

```python
tbl = ascii.read('./output/temp.cat')
```

```python
fig, ax = plt.subplots(1, 5, figsize = (20, 4))

ax[0].hist(-2.5 * np.log10(tbl['flux_4000'] * 1e-6 * 1e-23) - 48.6, bins = np.arange(18, 26, 0.2), edgecolor = 'black')
ax[0].set_xlabel('AB mag, 4000A', fontsize = 15)

ax[1].hist(-2.5 * np.log10(tbl['flux_5000'] * 1e-6 * 1e-23) - 48.6, bins = np.arange(18, 26, 0.2), edgecolor = 'black')
ax[1].set_xlabel('AB mag, 5000A', fontsize = 15)

ax[2].hist(-2.5 * np.log10(tbl['flux_6000'] * 1e-6 * 1e-23) - 48.6, bins = np.arange(18, 26, 0.2), edgecolor = 'black')
ax[2].set_xlabel('AB mag, 6000A', fontsize = 15)

ax[3].hist(-2.5 * np.log10(tbl['flux_7000'] * 1e-6 * 1e-23) - 48.6, bins = np.arange(18, 26, 0.2), edgecolor = 'black')
ax[3].set_xlabel('AB mag, 7000A', fontsize = 15)

ax[4].hist(-2.5 * np.log10(tbl['flux_8000'] * 1e-6 * 1e-23) - 48.6, bins = np.arange(18, 26, 0.2), edgecolor = 'black')
ax[4].set_xlabel('AB mag, 8000A', fontsize = 15)

fig.savefig('./figs/mag_distribution.png')
```

```python
fig, ax = plt.subplots(figsize = (5, 5))

ax.scatter(zs['z_spec'], zs['z_m1'], s = 10)
ax.plot([0, 2], [0, 2], color = 'black')

nmad = 1.48 * np.median(np.abs(zs['z_spec'] - zs['z_m1']))
ax.text(1.2, 1, 'NMAD = 'f'{nmad:.3f}', fontsize = 15)

ax.set_xlim(0, 2)
ax.set_ylim(0, 2)

ax.set_xlabel('spec-z', fontsize = 15)
ax.set_ylabel('phot-z', fontsize = 15)

fig.savefig('./figs/spec_phot_redshift.png')
```
